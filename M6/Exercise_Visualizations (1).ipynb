```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs


# Generating random data points
X, y = make_blobs(n_samples=30, centers=2, n_features=3, random_state=123)

# Storing data in a DataFrame
plot_data = pd.DataFrame(X, columns=['x1', 'x2', 'x3'])
```

```python
plot_data['x1'].plot(kind='line', colormap='viridis', title="Simple line plot")
plt.show()
```

```python
plot_data[['x2', 'x3']].plot(kind='bar', colormap='viridis', title="Bar plot for 2 variables")
plt.show()
```

```python
plot_data['x2'].plot(kind='hist', bins=5, colormap='viridis', title="Histogram")
plt.show()
```

```python
plot_data[['x2', 'x3']].plot.scatter(x='x2', y='x3', colormap='viridis', title="Scatter plot")
plt.show()
```

```python
ax = plot_data.boxplot()
plt.title("Boxplot for all 3 variables")
plt.show()
```
